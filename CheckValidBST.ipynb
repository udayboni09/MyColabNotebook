Check if the given tree is valid BST or not.
To determine a given tree is a valid BST or not, if value of left node is less than parent node & value of right not shouls always be greater than parent node then we call it is a valid BST

```python
#Step1: Defining the data structure for the Binary Search Tree

class Node(object):
  def __init__(self,val,left=None,right=None):
    self.val = val
    self.left = left
    self.right = right

#   5
# 4     7

#now we need to check if this a valid BST or not 
#Step2: Now we need to check whether BST is valid or not 

class Solution(object):

  def _isValidBSTHelper(self,n,low,high):
    #Base case if the node is empty then it is a valid BST
    if not n:
      return True 
    
    #Other than base case we need to recursivelly call the isValidBSTHelper with low and high values 
    val = n.val

    if (      (val > low and val < high) 
          and self._isValidBSTHelper(n.left,low,n.val) 
          and self._isValidBSTHelper(n.right,n.val,high)
          ):
      return True
    return False

  def isValidBST(self,n):
    return self._isValidBSTHelper(n,float('-inf'),float('inf'))
    
  
node1 = Node(5)
node1.left = Node(4)
node1.right = Node(7)


node2 = Node(5)
node2.left = Node(4)
node2.right = Node(7)
node2.right.left = Node(3)


print(Solution().isValidBST(node1))
print(Solution().isValidBST(node2))
```
