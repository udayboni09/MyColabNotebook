Following example shows DP Fibonacci Example Method 1

```python
def calculateFibonacci(n):
  #we will define a vector called momoization
  memoize = [-1 for each in range(n+1)]
  return calculateFibonacciRecrusive(n, memoize)

def calculateFibonacciRecrusive(n, memoize):
  #base case if the n < 2 return return n
  if n < 2:
    return n
  # if the sub problem is alrady solved then please get it from cache 
  if memoize[n] > 1:
    return memoize[n]
  #else calculate the subproblem 
  memoize[n] = calculateFibonacciRecrusive(n-1, memoize) + calculateFibonacciRecrusive(n-2, memoize)
  return  memoize[n]

def main():
  print("5th fibonacci is :", calculateFibonacci(5))
  print("6th fibonacci is :", calculateFibonacci(6))
  print("7th fibonacci is :", calculateFibonacci(7))
  print("10th fibonacci is :", calculateFibonacci(10))
  print("15th fibonacci is :", calculateFibonacci(15))
main()
```
