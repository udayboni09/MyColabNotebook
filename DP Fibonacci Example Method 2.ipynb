Fibonacci approach using botttom up approach of DP

```python
def calculateFibonacci(n):
  table = [0,1]

  for i in range(2, n+1):
    table.append(table[i-1] + table[i-2])
  
  return table[n]

def main():
  print("5th fibonacci is :", calculateFibonacci(5))
  print("6th fibonacci is :", calculateFibonacci(6))
  print("7th fibonacci is :", calculateFibonacci(7))
  print("10th fibonacci is :", calculateFibonacci(10))
  print("15th fibonacci is :", calculateFibonacci(15))
main()
```
