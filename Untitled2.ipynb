```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# No of Data points
N = 500
  
# initializing random values
data = [1, 1, 1, 2,5252,1123111]
```

```python
count, bins_count = np.histogram(data, bins=10)
```

```python
pdf = count / sum(count)
```

```python
pdf = count / sum(count)
cdf = np.cumsum(pdf)
```

```python
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```
