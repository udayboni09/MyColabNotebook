Find Two Numbers that Add up to "k" lst = [1,21,3,14,5,60,7,6]
k = 81

Approach:
Step1: First sort the array. We can use Mrge Sort as it is better to use because it's time complexity is n logn

Step2: After sorting create 2 indexes one points to 0th element and other one points to last element of the array 

Step3: Start traverssing from each side untill both are collidiaded. During this journey, check if arr[i] + arr[j] == k then simply cache them 


```python
def find_sum(arr,k):
  #base case:
  if len(arr) == 0 or k == 0:
    return "None" 
  i = 0
  j = len(arr) - 1
  #sort the array
  #o(n. login)
  arr.sort()
 
  sum = 0
  while (i != j):

    sum = arr[i] + arr[j]
    
    if sum < k:

      #we have to keep moving towards right side of the array 
      #increment i
      i += 1
     
    elif sum > k: 
      # we have to keep moving towards left side of the array 
      j -= 1
     
    else:

      #That means we found a place where sum is matching for a i and j
      #just cache them instead of new array, to avoid space complexity 
      result = "First Number: "+str(arr[i]) + " and Second Number is: " +str(arr[j])
      print(str)
      #print("First Number:",arr[i])
      #print("2nd Number:",arr[j])
      return  result 
      

  return "Nothing found"

def main():
  print(find_sum([1,21,3,14,5,56,7,6],81))

main()
```
